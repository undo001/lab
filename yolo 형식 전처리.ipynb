```python
import json
import os
```

```python
# JSON 파일이 있는 디렉토리 경로
json_dir = 'E:/2학기 양봉/106.지능형 양봉 데이터/22/106.지능형 양봉 데이터/01.데이터/Training/02.라벨링데이터/TL_01.실제데이터_01.Bounding Box_04.수일벌-이탈리안/'

# YOLO 형식의 TXT 파일로 저장할 디렉토리 경로
output_dir = 'E:/2학기 양봉/yolov8'

# 디렉토리가 존재하지 않으면 생성합니다.
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
# JSON 파일을 YOLO 형식의 텍스트 파일로 변환하고 저장합니다.
for json_filename in os.listdir(json_dir):
    if json_filename.endswith('.json'):
        with open(os.path.join(json_dir, json_filename), 'r') as json_file:
            data = json.load(json_file)

        # YOLO 형식의 텍스트 파일로 저장할 파일 경로
        output_file_path = os.path.join(output_dir, os.path.splitext(json_filename)[0] + '.txt')

        with open(output_file_path, 'w') as txt_file:
            for item in data['annotations']:
                class_id = 0  # 클래스 ID (수박 클래스의 경우 0으로 가정)
                x_center = (item['bbox']['bndex_xcrdnt'] + item['bbox']['bndex_width']/2) / ( data['image']['width'])
                y_center = (item['bbox']['bndex_ycrdnt'] + item['bbox']['bndex_hg']/2) / (data['image']['height'])
                width = item['bbox']['bndex_width'] / data['image']['width']
                height = item['bbox']['bndex_hg'] / data['image']['height']

                # YOLO 형식으로 변환된 라벨을 파일에 씁니다.
                txt_file.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

        print(f"JSON 파일 '{json_filename}'이 YOLO 형식의 라벨 파일로 저장되었습니다.")

print("모든 JSON 파일을 변환하여 저장했습니다.")
```

```python
import os

# 디렉토리 경로
디렉토리_경로 = 'C:/Users/lab/Desktop/파일/val'

# 디렉토리 내의 모든 파일 가져오기
file_list = os.listdir(디렉토리_경로)

# jpg 파일과 json 파일을 분리
jpg_files = [file for file in file_list if file.endswith('.jpg')]
json_files = [file for file in file_list if file.endswith('.txt')]

# jpg 파일 이름에서 확장자 제거하여 숫자만 남기기
jpg_numbers = [jpg_file.split('.jpg')[0] for jpg_file in jpg_files]

# 중복되지 않는 jpg 파일에 대해 연결된 json 파일을 삭제
for jpg_number in jpg_numbers:
    json_file = jpg_number + '.txt'
    if json_file not in json_files:
        jpg_file_path = os.path.join(디렉토리_경로, jpg_number + '.jpg')
        os.remove(jpg_file_path)

print("중복되지 않는 jpg 파일 중에 연결된 txt 파일이 없는 파일이 삭제되었습니다.")
```

```python
def convert_to_yolo_format(data):
    image_width = data["IMAGE"]["WIDTH"]
    image_height = data["IMAGE"]["HEIGHT"]
    image_file_name = data["IMAGE"]["IMAGE_FILE_NAME"]
    annotations = data["ANNOTATION_INFO"]
    
    yolo_labels = []
    for annotation in annotations:
        xtl = annotation["XTL"]
        ytl = annotation["YTL"]
        xbr = annotation["XBR"]
        ybr = annotation["YBR"]
        
        # Calculate normalized coordinates
        x_center = (xtl + xbr) / (2.0 * image_width)
        y_center = (ytl + ybr) / (2.0 * image_height)
        width = (xbr - xtl) / image_width
        height = (ybr - ytl) / image_height
        
        species = annotation["SPECIES"]
        lifecycle = annotation["LIFECYCLE"]
        
        yolo_label = f"{species} {x_center} {y_center} {width} {height} {lifecycle}\n"
        yolo_labels.append(yolo_label)
    
    return image_file_name, yolo_labels

# JSON 파일의 경로 지정
json_file_path = 'E:/2학기 양봉/yolov8/data/'

# JSON 파일 불러오기
with open(json_file_path, "r") as json_file:
    label_data = json.load(json_file)

# 출력 디렉토리 경로 지정
output_directory = 'E:/2학기 양봉/yolov8'

# JSON 데이터를 YOLO 형식으로 변환
image_file_name, yolo_labels = convert_to_yolo_format(label_data)

# 경로를 포함한 파일 이름 생성
yolo_filename = os.path.join(output_directory, os.path.splitext(image_file_name)[0] + ".txt")

# YOLO 라벨을 지정한 경로에 저장
with open(yolo_filename, "w") as file:
    file.writelines(yolo_labels)
```
